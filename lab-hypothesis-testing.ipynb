```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
alpha = 0.05
dragon_hp = df[df['Type 1'] == 'Dragon']['HP']
grass_hp = df[df['Type 1'] == 'Grass']['HP']

test_stat, p_value = st.ttest_ind(dragon_hp, grass_hp, alternative='greater', equal_var=True)
print(test_stat, p_value)


# Conclusion
if p_value < alpha:
    print("Reject the null hypothesis: Dragon-type Pokemon have, on average, more HP than Grass-type Pokemon.")
else:
    print("Fail to reject the null hypothesis: No significant difference in average HP between Dragon and Grass-type Pokemon.")
```

We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.

```python
legendary = df[df['Legendary'] == True]
non_legendary = df[df['Legendary'] == False]

stats_to_compare = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
results = {}

for stat in stats_to_compare:
    test_stat, p_value = st.ttest_ind(legendary[stat], non_legendary[stat], equal_var=True)
    results[stat] = p_value

adjusted_alpha = 0.05 / len(stats_to_compare)

for stat, p_value in results.items():
    if p_value < adjusted_alpha:
        print(f"Reject the null hypothesis for {stat}: Significant difference in {stat} between Legendary and Non-Legendary Pokémon (p-value: {p_value:.5f}).")
    else:
        print(f"Fail to reject the null hypothesis for {stat}: No significant difference in {stat} between Legendary and Non-Legendary Pokémon (p-value: {p_value:.5f}).")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

```python
alpha = 0.05
school_coords = np.array([-118, 37])
hospital_coords = np.array([-122, 34])

def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)

df['distance_to_school'] = euclidean_distance(df['longitude'], df['latitude'], *school_coords)
df['distance_to_hospital'] = euclidean_distance(df['longitude'], df['latitude'], *hospital_coords)

df['is_close'] = (df['distance_to_school'] < 0.50) | (df['distance_to_hospital'] < 0.50)

close_houses = df[df['is_close']].copy()
far_houses = df[~df['is_close']].copy()

test_stat, p_value = st.ttest_ind(close_houses['median_house_value'], far_houses['median_house_value'], equal_var=False)
print(test_stat, p_value)


# Conclusion
if p_value < alpha:
    print("Reject the null hypothesis: Houses close to a school or hospital are more expensive.")
else:
    print("Fail to reject the null hypothesis: No significant difference in price between houses close to and far from a school or hospital.")
```
